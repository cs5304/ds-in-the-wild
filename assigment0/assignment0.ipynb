```python
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)
xvals = np.linspace(0, 2*np.pi, 50)
yinterp = np.interp(xvals, x, y)
plt.plot(x, y, 'o')
plt.plot(xvals, yinterp, '-x')
plt.show()
```

```python
A, B, C = 10, 5, 15
s = torch.from_numpy(np.random.randn(A, A))
x = torch.from_numpy(np.random.randn(A, B))
y = torch.from_numpy(np.random.randn(A, C))
z = torch.matmul(x.t(), y)

print("""Tensor Statistics
=================
S-shape = {}
X-shape = {}
Y-shape = {}
Z-shape = {}
S-trace = {}
""".format(
    s.shape,
    x.shape,
    y.shape,
    z.shape,
    s.trace()
))
```
